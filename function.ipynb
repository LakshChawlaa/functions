```python
#Write a function to check whether a number is even or odd

def is_even_odd(num):
    if num %2==0:
        print("num is even")
    else:
        print("num is odd")

is_even_odd(24)       

```

```python
#Write a function that returns the maximum of two numbers.

def max(a,b):
    if a>b:
        print("a is max")
    else:
        print("b is max") 
max(5,6)        
```

```python
#write a factorial of the number 
def factorial(n):
    if n == 0:
        return 1 
    else:
        return n * factorial(n - 1)
print(factorial(5))
```

```python
#Write a function to check whether a number is prime.
def prime(a):
    if a<=1:
        return False 
    for i in range(2,a):
        if a % i == 0:
            return False 
    return True     
print(prime(33))
print(prime(5))        
```

```python
#Write a function that counts vowels in a string.
def vowels(b):
    vowels = "aeiouAEIOU"
    count = 0 
    for i in b:
        if i in vowels:
            count += 1
    return count  

print(vowels("hello world"))
```

```python
#Write a function to reverse a string.
def rev(s):
    return s [::-1]
print(rev("laksh"))
```

```python
#Write a function to find the sum of elements in a list.
def list(lst):
    total = 0 
    for i in lst:
        total +=  i 
    return total 
print(list([1,2,3,4,5,6,7]))      
```

```python
#Write a function to find the largest number in a list.
def large_ist(lst):
    large = lst[0]
    for i in lst:
        if i > large:
            large = i
    return large 
print(large_ist([4,9,80,7]))        
```

```python
#Write a function to remove duplicates from a list.
def duplicates(lst):
    result = []
    for i in lst:
        if i not in result:
            result.append(i)
    return result
print(duplicates([1, 2, 2, 3, 4, 4, 5,5,5,6,6,6,7,7]))

    
```

```python
#Write a function that returns only even numbers from a list
def even(lst):
    even = []
    for i in lst:
        if i%2==0:
            even.append(i)
    return even
print(even([1,2,3,4,5,6,7,8]))        
```
